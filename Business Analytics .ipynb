```python
df['Date'] = pd.to_datetime(
    df['Date'],
    errors='coerce',
    dayfirst=True
)
```

```python
df['Date'].isna().sum()
df.info()
```

```python
df['Business_Unit'] = (df['Business_Unit'].str.strip().str.title().fillna('Unknown'))
df['Region'] = (df['Region'].str.strip().str.title().fillna('Unknown'))
```

```python
df.head()
```

```python
import numpy as np
```

```python
numeric_cols = [
    'Revenue', 'Cost', 'Operating_Expense',
    'Employee_Count', 'Customer_Count',
    'Marketing_Spend', 'Production_Volume'
]
for col in numeric_cols:
    df[col] = df[col].apply(lambda x: x if x >= 0 else np.nan)
```

```python
df.loc[df['Attrition_Count'] > df['Employee_Count'], 'Attrition_Count'] = np.nan
df.loc[df['Customer_Churn'] > df['Customer_Count'], 'Customer_Churn'] = np.nan
```

```python
df['Revenue'] = df['Revenue'].fillna(df['Revenue'].median())
```

```python
df['Revenue'].isna().sum()
df['Revenue'].describe()
```

```python
df['Cost'] = df['Cost'].fillna(df['Cost'].median())
df['Operating_Expense'] = df['Operating_Expense'].fillna(df['Operating_Expense'].median())
df['Employee_Count'] = df['Employee_Count'].fillna(df['Employee_Count'].median())
df['Customer_Count'] = df['Customer_Count'].fillna(df['Customer_Count'].median())
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

```python
df['Marketing_Spend'] = df['Marketing_Spend'].fillna(0)
```

```python
df['Production_Volume'] = df['Production_Volume'].fillna(
    df['Production_Volume'].median())
```

```python
df.to_csv("C:/Users/Test/Desktop/Cleaned_Business_Performance_Dataset.csv", index=False)
```
